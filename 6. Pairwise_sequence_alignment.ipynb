```python
from Bio import Align
Aligner = Align.PairwiseAligner(match_score= 1, mismatch_score=0)
```

```python
#defining to sequences for alignment

SeqA="GAACTCGTTCGATGA"
SeqB="ATTGCTACGC"

#call the aligner to assign score
score= Aligner.score(SeqA, SeqB)
print(score)
```

```python
#for GLOBAL allignment

Alignments = Aligner.align(SeqA, SeqB)
for alignment in Alignments:
  print(alignment)
```

```python
#For LOCAL allignment
Align.mode="local"
Seq1="GAACTCGTTCGATGA"
Seq2="ATTGCCGC"

#call the aligner to assign score
score= Aligner.score(Seq1, Seq2)
print(score)
```

```python
Aligner.mode = 'local'
Alignments = Aligner.align(Seq1, Seq2)
for alignment in Alignments:
  print(alignment)
```

```python
Align.mode="local"
Seq3="GAAAAAACTCGTTCGATGA"
Seq4="AAAAAAAATTGCCGC"

score= Aligner.score(Seq3, Seq4)
print(score)
```
