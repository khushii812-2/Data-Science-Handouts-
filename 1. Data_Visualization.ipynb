```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,6,8]

c=['r','y','b','g']
plt.bar(x,y,color=c)

plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,6,8]

c=['r','y','b','g']
plt.scatter(x,y,color=c)

plt.show()
```
